```python
# =========================================================
# 1. IMPORT LIBRARIES
# =========================================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder, PolynomialFeatures
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.manifold import TSNE

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier

# =========================================================
# 2. LOAD DATASET
# =========================================================
path = "/kaggle/input/crop-recommendation-dataset/Crop_recommendation.csv"
df = pd.read_csv(path)

# =========================================================
# 3. PREPROCESSING (4)
# =========================================================
df.fillna(df.median(numeric_only=True), inplace=True)

def iqr_clip(data, cols):
    for col in cols:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        data[col] = np.clip(data[col], Q1 - 1.5*IQR, Q3 + 1.5*IQR)
    return data

num_cols = ['N','P','K','temperature','humidity','ph','rainfall']
df = iqr_clip(df, num_cols)

le = LabelEncoder()
df['label'] = le.fit_transform(df['label'])

# =========================================================
# 4. FEATURE ENGINEERING (4)
# =========================================================
df['N_P_ratio'] = df['N'] / (df['P'] + 1)
df['P_K_ratio'] = df['P'] / (df['K'] + 1)
df['N_K_ratio'] = df['N'] / (df['K'] + 1)

df['NPK_mean'] = df[['N','P','K']].mean(axis=1)
df['NPK_std']  = df[['N','P','K']].std(axis=1)

df['rainfall_log'] = np.log1p(df['rainfall'])
df['humidity_log'] = np.log1p(df['humidity'])

X_base = df.drop('label', axis=1)
y = df['label']

poly = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)
X_poly = poly.fit_transform(X_base)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_poly)

# =========================================================
# 5. TRAIN TEST SPLIT
# =========================================================
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

# =========================================================
# 6. MODELS (6)
# =========================================================
models = {
    "Logistic Regression": LogisticRegression(max_iter=3000),
    "KNN": KNeighborsClassifier(n_neighbors=7),
    "SVM": SVC(kernel='rbf'),
    "Random Forest": RandomForestClassifier(n_estimators=300, max_depth=20, random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=300),
    "XGBoost": XGBClassifier(
        n_estimators=300, learning_rate=0.05,
        max_depth=6, eval_metric='mlogloss'
    )
}

# =========================================================
# 7. TRAINING + EVALUATION
# =========================================================
results = []

for name, model in models.items():
    print(f"\n================ {name} =================")

    model.fit(X_train, y_train)
    y_test_pred = model.predict(X_test)

    train_acc = accuracy_score(y_train, model.predict(X_train))
    test_acc  = accuracy_score(y_test, y_test_pred)
    precision = precision_score(y_test, y_test_pred, average='weighted')
    recall    = recall_score(y_test, y_test_pred, average='weighted')
    f1        = f1_score(y_test, y_test_pred, average='weighted')

    print("Train Acc:", train_acc)
    print("Test Acc :", test_acc)
    print(classification_report(y_test, y_test_pred))

    # -------- Confusion Matrix (WITH VALUES) --------
    cm = confusion_matrix(y_test, y_test_pred)
    plt.figure(figsize=(6,5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f"Confusion Matrix – {name}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

    # -------- t-SNE FOR THIS MODEL --------
    tsne = TSNE(n_components=2, perplexity=40, random_state=42)
    X_test_tsne = tsne.fit_transform(X_test)

    plt.figure(figsize=(7,6))
    sns.scatterplot(
        x=X_test_tsne[:,0],
        y=X_test_tsne[:,1],
        hue=y_test,
        palette='tab20',
        legend=False
    )
    plt.title(f"t-SNE Visualization – {name}")
    plt.xlabel("Component 1")
    plt.ylabel("Component 2")
    plt.show()

    results.append([name, train_acc, test_acc, precision, recall, f1])

# =========================================================
# 8. FINAL RESULT TABLE
# =========================================================
results_df = pd.DataFrame(
    results,
    columns=["Model", "Train Acc", "Test Acc", "Precision", "Recall", "F1-score"]
)

print("\nFINAL COMPARISON TABLE\n")
print(results_df)
```

```python
# =========================================================
# 1. IMPORT LIBRARIES
# =========================================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder, PolynomialFeatures
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.manifold import TSNE

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier

# =========================================================
# 2. LOAD DATASET
# =========================================================
path = "/kaggle/input/crop-recommendation-dataset/Crop_recommendation.csv"
df = pd.read_csv(path)

# =========================================================
# 3. PREPROCESSING (4)
# =========================================================
# (1) Missing value handling
df.fillna(df.median(numeric_only=True), inplace=True)

# (2) Outlier handling using IQR
def iqr_clip(data, cols):
    for col in cols:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        data[col] = np.clip(data[col], Q1 - 1.5*IQR, Q3 + 1.5*IQR)
    return data

num_cols = ['N','P','K','temperature','humidity','ph','rainfall']
df = iqr_clip(df, num_cols)

# (3) Label Encoding
le = LabelEncoder()
df['label'] = le.fit_transform(df['label'])
class_names = le.classes_   # <-- for t-SNE legend

# =========================================================
# 4. FEATURE ENGINEERING (4)
# =========================================================
# (1) Ratio features
df['N_P_ratio'] = df['N'] / (df['P'] + 1)
df['P_K_ratio'] = df['P'] / (df['K'] + 1)
df['N_K_ratio'] = df['N'] / (df['K'] + 1)

# (2) Statistical features
df['NPK_mean'] = df[['N','P','K']].mean(axis=1)
df['NPK_std']  = df[['N','P','K']].std(axis=1)

# (3) Log transformation
df['rainfall_log'] = np.log1p(df['rainfall'])
df['humidity_log'] = np.log1p(df['humidity'])

# (4) Polynomial interaction features
X_base = df.drop('label', axis=1)
y = df['label']

poly = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)
X_poly = poly.fit_transform(X_base)

# =========================================================
# 5. SCALING
# =========================================================
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_poly)

# =========================================================
# 6. TRAIN TEST SPLIT
# =========================================================
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

# =========================================================
# 7. MODELS (6)
# =========================================================
models = {
    "Logistic Regression": LogisticRegression(max_iter=3000),
    "KNN": KNeighborsClassifier(n_neighbors=7),
    "SVM (RBF)": SVC(kernel='rbf'),
    "Random Forest": RandomForestClassifier(
        n_estimators=300, max_depth=20, random_state=42
    ),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=300),
    "XGBoost": XGBClassifier(
        n_estimators=300,
        learning_rate=0.05,
        max_depth=6,
        eval_metric='mlogloss'
    )
}

# =========================================================
# 8. TRAINING + EVALUATION
# =========================================================
results = []

for name, model in models.items():
    print(f"\n================ {name} =================")

    model.fit(X_train, y_train)
    y_test_pred = model.predict(X_test)

    train_acc = accuracy_score(y_train, model.predict(X_train))
    test_acc  = accuracy_score(y_test, y_test_pred)
    precision = precision_score(y_test, y_test_pred, average='weighted')
    recall    = recall_score(y_test, y_test_pred, average='weighted')
    f1        = f1_score(y_test, y_test_pred, average='weighted')

    print("Train Accuracy:", train_acc)
    print("Test Accuracy :", test_acc)
    print("\nClassification Report:\n")
    print(classification_report(y_test, y_test_pred, target_names=class_names))

    # ---------------- Confusion Matrix ----------------
    cm = confusion_matrix(y_test, y_test_pred)
    plt.figure(figsize=(7,6))
    sns.heatmap(
        cm, annot=True, fmt='d', cmap='Blues',
        xticklabels=class_names,
        yticklabels=class_names
    )
    plt.title(f"Confusion Matrix – {name}")
    plt.xlabel("Predicted Class")
    plt.ylabel("True Class")
    plt.xticks(rotation=90)
    plt.yticks(rotation=0)
    plt.show()

    # ---------------- t-SNE with Color–Class Meaning ----------------
    tsne = TSNE(n_components=2, perplexity=40, random_state=42)
    X_test_tsne = tsne.fit_transform(X_test)

    plt.figure(figsize=(8,7))
    sns.scatterplot(
        x=X_test_tsne[:,0],
        y=X_test_tsne[:,1],
        hue=y_test,
        palette='tab20',
        legend='full'
    )

    handles, _ = plt.gca().get_legend_handles_labels()
    plt.legend(
        handles,
        class_names,
        title="Crop Classes",
        bbox_to_anchor=(1.05, 1),
        loc='upper left'
    )

    plt.title(f"t-SNE Visualization – {name}")
    plt.xlabel("Component 1")
    plt.ylabel("Component 2")
    plt.tight_layout()
    plt.show()

    results.append([name, train_acc, test_acc, precision, recall, f1])

# =========================================================
# 9. FINAL RESULT TABLE
# =========================================================
results_df = pd.DataFrame(
    results,
    columns=["Model", "Train Acc", "Test Acc", "Precision", "Recall", "F1-score"]
)

print("\nFINAL MODEL COMPARISON TABLE\n")
print(results_df)
```
